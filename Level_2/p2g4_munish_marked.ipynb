```python
# Missing data exploration
def missing_data_in_dataframe(input_df):
    df_missing = input_df.isnull().sum().sort_values(ascending = False)
    df_missing = df_missing.reset_index()
    df_missing['Percent'] = df_missing.iloc[:, 1].apply(lambda x: x*100/np.sum(df_missing.iloc[:, 1]))
    df_missing.columns = ['Attributes', 'Missing', 'Percent']
    mask = df_missing['Missing'] > 0
    df_missing = df_missing[mask]
    return df_missing
```

```python
df_missing_data = missing_data_in_dataframe(df_ames_raw)
print(df_missing_data) #Print Missing Data

#Plot Missing Data
fig, ax = plt.subplots(figsize=(15,10))  # Set size of plot Create figure.
sns.barplot(x = 'Percent', y = 'Attributes', data = df_missing_data, palette = 'viridis', ax=ax)
```

```python
def split_dataframe_data(input_df, input_target='SalePrice', input_source='train', input_test_size=0.3):
    X = input_df.drop([input_target], axis = 1).values
    y = np.log10(input_df[input_target].values)
    
    # Split the df_ames_cleaned dataset into train and test sets (default : 70:30 split)
    out_X_train, out_X_test, out_y_train, out_y_test = train_test_split(
                                                        X, 
                                                        y, 
                                                        test_size = input_test_size, 
                                                        random_state = 42)
    
    return out_X_train, out_X_test, out_y_train, out_y_test
```

```python
def plot_heatmap_for_a_corr(input_cor):

    # Set the default matplotlib figure size :
    fig, ax = plt.subplots(figsize=(18,12))         # Sample figsize in inches

    # Generate a mask for the upper triangle (taken from seaborn example gallery)
    mask = np.zeros_like(input_cor, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Plot the heatmap with seaborn.
    # Assign the matplotlib axis the function returns. This will let us resize the labels.
    ax = sns.heatmap(input_cor, mask=mask, ax=ax, annot=True, cmap='coolwarm', linewidths=.5)

    # Resize the labels.
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=10)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=12)

    # If you put plt.show() at the bottom, it prevents those useless printouts from matplotlib.
    plt.show()
    return
```

```python
#Plotting for Numeric Data
def subplot_distplot(dataframe,columns): 
    nrows = int(np.ceil(len(columns)/3)) # Allocate enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=3, figsize=(12,nrows*4))
    ax = ax.ravel()
    for i, column in enumerate(columns):
        sns.distplot(dataframe[column],ax=ax[i])
        
        ylim = ax[i].get_ylim()[1]
        ax[i].set_title('Distribution of '+column)
        ax[i].set_ylabel('kde')
    plt.tight_layout()
```

```python
#Get variables which have high correlation to target
def get_high_corr(df,target,corr,corrmethod='pearson'): 
    return [i for i in df.corr(method=corrmethod).index 
            if abs(df.corr(method=corrmethod).loc[i,target])>corr and (i!=target)]
```

```python
def get_collinearity(df,target,corr,corrmethod='pearson'): #define function to get list of variables with possibly collinearity
    df = df.drop(target,axis=1)
    variables = set([tuple(sorted([i,j])) 
                     for i in df.corr(method=corrmethod).index 
                     for j in df.corr(method=corrmethod).columns 
                     if (abs(df.corr(method=corrmethod).loc[i,j]>corr)) and (i!=j)])
    x_var = [var[0] for var in variables]
    y_var = [var[1] for var in variables]
    return x_var, y_var
```

```python
def subplot_scatter(dataframe,cols_x,cols_y): #plot multiple scatter plots
    nrows = int(np.ceil(len(cols_x)/3)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=3, figsize=(12,nrows*4))
    ax = ax.ravel()
    for i, column in enumerate(cols_x):
        sns.scatterplot(x=dataframe[column],y=dataframe[cols_y[i]],ax=ax[i])
        ax[i].set_title(cols_y[i]+' vs '+column)
    plt.tight_layout()
```
