## 2. Remove non-numeric columns whose variation of data is minimal 
(ie. too many of the same category/value/type. threshold = 85%)

```python
# Get numeric columns
numeric_cols = df._get_numeric_data().columns

# Get non-numeric columns
non_numeric_cols = [i for i in df.columns if i not in numeric_cols]
```

```python
little_info_cols = []

for i in non_numeric_cols:
    
    # Get most no of occurence of a value in the column
    max_cnts = df[i].value_counts().max()
    
    # Get the total number of non-null values in the column
    sum_ = df[i].value_counts().sum()
    
    # If the percentage of max occurence > 85%
    ptg_of_data = max_cnts/sum_
    if ptg_of_data > 0.85:
        little_info_cols.append(i)

little_info_cols
```

```python
# Drop them
df.drop(columns = little_info_cols, inplace=True)
```

## 3. Remove NUMERIC columns whose variation of data is minimal 
(i.e. too many of the same category/value/type. threshold = 85%)

```python
little_info_cols = []

for i in numeric_cols:
    
    # Get most no of occurence of a value in the column
    max_cnts = df[i].value_counts().max()
    
    # Get the number of non-null values in the column
    sum_ = df[i].value_counts().sum()
    
    # If the percentage of max occurence > 85%
    ptg_of_data = max_cnts/sum_
    if ptg_of_data > 0.85:
        little_info_cols.append(i)

little_info_cols
```

```python
# Drop them
df.drop(little_info_cols, axis=1, inplace=True)
```

### `BsmtFin SF 1` + `Bsmt Unf SF` = `Total Bsmt SF	`

```python
df['checks'] = df.apply(lambda x: 1 if (x['BsmtFin SF 1'] + x['Bsmt Unf SF']) == x['Total Bsmt SF'] else 0, axis=1)
```

```python
round(len(df[df.checks == 1]) / len(df) * 100, 1)
```

- *Even though some rows have sum not equals to 'Total Bsmt SF', 87%. of them are. We decide keep 'Total Bsmt SF' being more holistic, and only 'BsmtFin SF 1'*

```python
df.drop(columns='Bsmt Unf SF', inplace=True)
```

### `1st Flr SF` + `2nd Flr SF` = `Gr Liv Area`

```python
df['checks'] = df.apply(lambda x: 1 if (x['1st Flr SF'] + x['2nd Flr SF']) == x['Gr Liv Area'] else 0, axis=1)
```

```python
round(len(df[df.checks == 1]) / len(df) * 100, 1)
```

- *Even though there are rows with sum not equals to 'Gr Liv Area', 98.8%. of them are. We decide to drop '2nd Flr SF' to remove its explicit relationship with 'Gr Liv Area'.*

```python
df.drop(columns='2nd Flr SF', inplace=True)
```
